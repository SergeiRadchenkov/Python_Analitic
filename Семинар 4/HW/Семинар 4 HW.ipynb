### 1.1. Скачать следующие данные: kc-house-data и laptop_price

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
houses = pd.read_csv('kc_house_data.csv')
laptops = pd.read_csv('laptop_price.csv', encoding='windows-1251')
```

### 1.2. Изучите стоимости недвижимости

```python
plt.figure(figsize=(12, 6))
max_x = np.max(np.arange(250000, 10000000, 250000))
bins = np.arange(0, max_x, 250000)
plt.hist(houses['price'], bins = bins)
plt.title('Cтоимости недвижимости')
plt.xlabel('Цена, тыс.')
plt.ylabel('Количество')
plt.xticks(np.arange(250000, 10000000, 250000), [f'{int(x / 1000)}' for x in np.arange(250000, 10000000, 250000)], rotation=45)
plt.yticks(np.arange(0, 10000, 1000))
plt.grid();
```

#### Выводы:
- больше всего домов с ценой от 250 тыс. до 500 тыс. - более 9000.
- домов со стоимостью менее 250 тыс. в два раза меньше, чем домов от 250 тыс. до 500 тыс.
- количество домов значитально уменьшается с ростом цены от 500 тыс.
- возможно присутствуют выбросы

### 1.3. Изучите распределение квадратуры жилой

```python
plt.figure(figsize=(12, 6))
max_x = np.max(np.arange(500, 15000, 500))
bins = np.arange(0, max_x, 500)
plt.hist(houses['sqft_living'], bins = bins)
plt.title('Распределение квадратуры жилой')
plt.xlabel('м.кв')
plt.ylabel('Количество')
plt.xticks(np.arange(500, 15000, 500), rotation=42)
plt.yticks(np.arange(0, 5500, 500))
plt.grid();
```

#### Выводы:
- больше всего домов с жилой площадью от 1500 м.кв до 2000 м.кв - более 5000.
- домов с жилой площадью менее 500 м.кв незначительное количество.
- количество домов значитально уменьшается с ростом жилой площади от 2500 м.кв.
- возможно присутствуют выбросы

### 1.4. Изучите распределение года постройки

```python
plt.figure(figsize=(16, 6))
plt.hist(houses['yr_built'], bins=23)
plt.title('Распределение годов постройки домов')
plt.xlabel('Год постройки')
plt.ylabel('Количество домов')
plt.xticks(range(1900, 2020, 5), rotation=42)
plt.grid()
```

#### Выводы:
- c 1900 года по 2015 год наблюдается рост количества построенных домов
- присутствуют 4 периода падения количества построенных домов:
1. С 1930 по 1940 года.
2. С 1970 по 1975 года
3. С 1980 по 2000 наблюдается небольшой застой.
4. С 2010 по 2015 года.

### 2.1. Изучите распределение домов от наличия вида на набережную

```python
ocean_view = houses['waterfront'].value_counts()
ocean_view
```

```python
plt.figure(figsize=(16, 6))
plt.title('Распределение домов от наличия вида на набережную')
ocean_view.index = ['Дома без вида на набережную', 'Дома с видом на набережную']
plt.pie(ocean_view, labels=ocean_view.index, autopct='%.1f');
```

#### Выводы:
- домов с видом на набережную незначительное количество от общего числа домов

### 2.2. Изучите распределение этажей домов

```python
floors = houses['floors'].value_counts()
floors
```

```python
plt.figure(figsize=(16, 6))
plt.title('Распределение этажей домов')
plt.xlabel('Этажи')
plt.ylabel('Количество')
plt.xticks(rotation=35)
plt.grid(axis='y')
plt.bar(floors.index, floors, width=0.4);
```

#### Выводы:
- больше 10 тыс. домов одноэтажной постройки.
- двухэтажные дома на втором месте - более 8 тыс.
- остальные количества незначительные.
максимальное количество этажей - 3,5.
